```python
from efficient_probit_regression.sampling import calculate_lewis_weights_exact, calculate_l2_lp_leverage_score, to_density

# 1. Laden der notwendigen Datensätze
%store -r Iris
%store -r Webspam
%store -r KDDCup
%store -r Covertype
%store -r Example2D

#Untersuchung für unterschiedliche Werte von p
p = [1.0, 1.3, 1.5, 1.7, 2.0]

for i in p:
    print("p = ", i)
    
    #Iris Datensatz
    print("Iris Datensatz")
    Iris_lewis_weights = calculate_lewis_weights_exact(Iris.X, p=i, T=20)
    Iris_lewis_prob = to_density(Iris_lewis_weights)
    Iris_l2_lp_scores = calculate_l2_lp_leverage_score(Iris.X, p=i)[0]
    Iris_l2_lp_prob = calculate_l2_lp_leverage_score(Iris.X, p=i)[1]
    Iris_c = Iris_lewis_prob / Iris_l2_lp_prob
    print("Iris_c:")
    print(Iris_c)
    
    #Example2D Datensatz
    print("Example2D Datensatz")
    Example2D_lewis_weights = calculate_lewis_weights_exact(Example2D.X, p=i, T=20)
    Example2D_lewis_prob = to_density(Example2D_lewis_weights)
    Example2D_l2_lp_scores = calculate_l2_lp_leverage_score(Example2D.X, p=i)[0]
    Example2D_l2_lp_prob = calculate_l2_lp_leverage_score(Example2D.X, p=i)[1]
    Example2D_c = Example2D_lewis_prob / Example2D_l2_lp_prob
    print("Example2D_c:")
    print(Example2D_c)

    #KDDCup Datensatz
    print("KDDCup Datensatz")
    KDDCup_lewis_weights = calculate_lewis_weights_exact(KDDCup.X, p=i, T=20)
    KDDCup_lewis_prob = to_density(KDDCup_lewis_weights)
    KDDCup_l2_lp_scores = calculate_l2_lp_leverage_score(KDDCup.X, p=i)[0]
    KDDCup_l2_lp_prob = calculate_l2_lp_leverage_score(KDDCup.X, p=i)[1]
    KDDCup_c = KDDCup_lewis_prob / KDDCup_l2_lp_prob
    print("KDDCup_c[1:100]:")
    print(KDDCup_c[0:100])
    print("KDDCup_c[-100:]:")
    print(KDDCup_c[-100:])

    #Covertype Datensatz
    print("Covertype Datensatz")
    Covertype_lewis_weights = calculate_lewis_weights_exact(Covertype.X, p=i, T=20)
    Covertype_lewis_prob = to_density(Covertype_lewis_weights)
    Covertype_l2_lp_scores = calculate_l2_lp_leverage_score(Covertype.X, p=i)[0]
    Covertype_l2_lp_prob = calculate_l2_lp_leverage_score(Covertype.X, p=i)[1]
    Covertype_c = Covertype_lewis_prob / Covertype_l2_lp_prob
    print("Covertype_c[1:100]:")
    print(Covertype_c[0:100])
    print("Covertype_c[-100:]:")
    print(Covertype_c[-100:])

    #Webspam Datensatz
    print("Webspam Datensatz")
    Webspam_lewis_weights = calculate_lewis_weights_exact(Webspam.X, p=i, T=20)
    Webspam_lewis_prob = to_density(Webspam_lewis_weights)
    Webspam_l2_lp_scores = calculate_l2_lp_leverage_score(Webspam.X, p=i)[0]
    Webspam_l2_lp_prob = calculate_l2_lp_leverage_score(Webspam.X, p=i)[1]
    Webspam_c = Webspam_lewis_prob / Webspam_l2_lp_prob
    print("Webspam_c[1:100]:")
    print(Webspam_c[0:100])
    print("Webspam_c[-100:]:")
    print(Webspam_c[-100:])
```
