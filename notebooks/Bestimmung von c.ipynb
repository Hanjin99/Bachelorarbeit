```python
import numpy as np

from efficient_probit_regression.datasets import Covertype
from efficient_probit_regression.sampling import calculate_lewis_weights_exact, calculate_l2_lp_leverage_score, to_density, compute_leverage_scores, calculate_lev_score_exact

%store -r Covertype

print("Covertype Datensatz")
Covertype_lewis_weights = calculate_lewis_weights_exact(Covertype.X, p=2, T=20)
Covertype_lewis_prob = to_density(Covertype_lewis_weights)
Covertype_l2_lp_scores = calculate_l2_lp_leverage_score(Covertype.X, p=2)[0]
Covertype_l2_lp_prob = calculate_l2_lp_leverage_score(Covertype.X, p=2)[1]
c_max = np.maximum(Covertype_lewis_prob / Covertype_l2_lp_prob, Covertype_l2_lp_prob / Covertype_lewis_prob)
Covertype_c = max(c_max)
print("Covertype_c:", Covertype_c)
print("")


index_max = np.argmax(c_max)
print("Das Maximum von c befindet sich am Index:", index_max)
print("Der Wert von c am Index:", c_max[index_max])
print("Covertype_lewis_prob[index_max]:", Covertype_lewis_prob[index_max])
print("Covertype_l2_lp_prob[index_max]:", Covertype_l2_lp_prob[index_max])
print("")

print("Nach der Untersuchung ist bei Covertype die Scores von lp und l2 unterschiedlich, obwohl p=2 gilt.")
print("")

test_lp = compute_leverage_scores(Covertype.X, p=2)
test_l2 = calculate_lev_score_exact(Covertype.X)
print("Rang von X:", Covertype.X.shape[1])
print("Summe von lp-Scores:", sum(test_lp))
print("Summe von l2-Scores:", sum(test_l2))
print("Hier sieht man, dass die Summe von l2-Scores nicht gleich dem vollen Rang von X ist.")
print("Die Funktion np.linalg.pinv() macht intern eine Singulärwertzerlegung (SVD) und schneidet sehr kleine Singulärwerte ab.")
print("Dadurch werden 2 Richtungen als praktisch 0 behandelt. Deshalb wird der effektive numerische Rang zu 53.")
print("Dies deutet darauf hin, dass X (oder XtX) stark schlecht konditioniert ist.")
print("")
```

```python
import numpy as np
import pandas as pd
from efficient_probit_regression.sampling import calculate_lewis_weights_exact, compute_leverage_scores, to_density

# Laden der notwendigen Datensätze
%store -r Iris
%store -r Webspam
%store -r KDDCup
%store -r Covertype
%store -r Example2D

print("Vergleich von min(l_ip, l_i2) * 0.5 <= w_i <= max(l_ip, l_i2) * 2:")
print("")

p = [1.0, 1.3, 1.5, 1.7, 2.0]
rows_cmp = []

datasets = {
    "Iris": Iris,
    "Example2D": Example2D,
    "Webspam": Webspam,
    "Covertype": Covertype,
    "KDDCup": KDDCup,
}

def process_dataset(dataset_name, X, p_val, rows_cmp, T=20):
    print(f"{dataset_name} Datensatz")
    print("")

    # l2 und lp leverage scores + Dichten
    l2_scores = compute_leverage_scores(X, p=2.0)
    l2_prob = to_density(l2_scores)

    lp_scores = compute_leverage_scores(X, p=p_val)
    lp_prob = to_density(lp_scores)

    # Lewis weights + Dichte
    lewis_weights = calculate_lewis_weights_exact(X, p=p_val, T=T)
    lewis_prob = to_density(lewis_weights)

    # Vergleich von Scores
    print("Vergleich von Scores")
    min_scores = np.minimum(lp_scores, l2_scores)
    max_scores = np.maximum(lp_scores, l2_scores)
    compare_scores = (lewis_weights >= min_scores * 0.5) & (lewis_weights <= max_scores * 2)

    print("Anzahl von TRUE:", int(np.sum(compare_scores)),
          "Anzahl von FALSE:", int(len(compare_scores) - np.sum(compare_scores)))
    print(round(np.sum(compare_scores) / len(compare_scores), 4) * 100, "% der Werte sind TRUE")
    print("")

    # Vergleich von normierten Verteilungen
    print("Vergleich von normierten Verteilungen")
    min_prob = np.minimum(lp_prob, l2_prob)
    max_prob = np.maximum(lp_prob, l2_prob)
    compare_prob = (lewis_prob >= min_prob * 0.5) & (lewis_prob <= max_prob * 2)

    # ---- Speichern (gleich wie vorher) ----
    n = len(compare_scores)

    true_scores = int(np.sum(compare_scores))
    false_scores = int(n - true_scores)
    ratio_true_scores = true_scores / n

    true_prob = int(np.sum(compare_prob))
    false_prob = int(n - true_prob)
    ratio_true_prob = true_prob / n

    rows_cmp.append({
        "dataset": dataset_name,
        "p": float(p_val),
        "n": int(n),
        "true_scores": true_scores,
        "false_scores": false_scores,
        "ratio_true_scores": ratio_true_scores,
        "true_prob": true_prob,
        "false_prob": false_prob,
        "ratio_true_prob": ratio_true_prob,
    })

    print("Anzahl von TRUE:", int(np.sum(compare_prob)),
          "Anzahl von FALSE:", int(len(compare_prob) - np.sum(compare_prob)))
    print(round(np.sum(compare_prob) / len(compare_prob), 4) * 100, "% der Werte sind TRUE")
    print("")


for p_val in p:
    print("p = ", p_val)
    print("")
    for dataset_name, ds in datasets.items():
        process_dataset(dataset_name, ds.X, p_val, rows_cmp, T=20)
```

```python
# Als csv speichern
df_cmp = pd.DataFrame(rows_cmp).sort_values(["dataset", "p"]).reset_index(drop=True)

float_cols = df_cmp.select_dtypes(include=["float"]).columns
df_cmp[float_cols] = df_cmp[float_cols].round(4)

out_csv = "Tabellen/Vergleich_minmax_l2lp_vs_lewis.csv"
df_cmp.to_csv(out_csv, index=False, float_format="%.4f")
print(f"CSV gespeichert: {out_csv}")
```

```python
p = 1.0

#Webspam Datensatz
print("Webspam Datensatz")
print("")
Webspam_l2_scores = compute_leverage_scores(Webspam.X, p=2.0)
Webspam_l2_prob = to_density(Webspam_l2_scores)
Webspam_lp_scores = compute_leverage_scores(Webspam.X, p=p)
Webspam_lp_prob = to_density(Webspam_lp_scores)
Webspam_lewis_weights = calculate_lewis_weights_exact(Webspam.X, p=p, T=20)
Webspam_lewis_prob = to_density(Webspam_lewis_weights)

# Vergleich von Scores
print("Vergleich von Scores")
Webspam_min_l2_lp_scores = np.minimum(Webspam_lp_scores, Webspam_l2_scores)
Webspam_max_l2_lp_scores = np.maximum(Webspam_lp_scores, Webspam_l2_scores)
Webspam_compare_scores = (Webspam_lewis_weights >= Webspam_min_l2_lp_scores * 0.5) & (
            Webspam_lewis_weights <= Webspam_max_l2_lp_scores * 2)

print("Anzahl von TRUE:", sum(Webspam_compare_scores), "Anzahl von FALSE:",
      len(Webspam_compare_scores) - sum(Webspam_compare_scores))
print(round(sum(Webspam_compare_scores) / len(Webspam_compare_scores), 4) * 100, "% der Werte sind TRUE")
print("")

# Vergleich von normierten Verteilungen
print("Vergleich von normierten Verteilungen")
Webspam_min_l2_lp_prob = np.minimum(Webspam_lp_prob, Webspam_l2_prob)
Webspam_max_l2_lp_prob = np.maximum(Webspam_lp_prob, Webspam_l2_prob)
Webspam_compare_prob = (Webspam_lewis_prob >= Webspam_min_l2_lp_prob * 0.5) & (Webspam_lewis_prob <= Webspam_max_l2_lp_prob * 2)

print("Anzahl von TRUE:", sum(Webspam_compare_prob), "Anzahl von FALSE:", len(Webspam_compare_prob) - sum(Webspam_compare_prob))
print(round(sum(Webspam_compare_prob) / len(Webspam_compare_prob), 4) * 100, "% der Werte sind TRUE")

print("")
```
