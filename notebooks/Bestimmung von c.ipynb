```python
import numpy as np

from efficient_probit_regression.datasets import Covertype
from efficient_probit_regression.sampling import calculate_lewis_weights_exact, calculate_l2_lp_leverage_score, to_density, compute_leverage_scores, calculate_lev_score_exact

%store -r Covertype

print("Covertype Datensatz")
Covertype_lewis_weights = calculate_lewis_weights_exact(Covertype.X, p=2, T=20)
Covertype_lewis_prob = to_density(Covertype_lewis_weights)
Covertype_l2_lp_scores = calculate_l2_lp_leverage_score(Covertype.X, p=2)[0]
Covertype_l2_lp_prob = calculate_l2_lp_leverage_score(Covertype.X, p=2)[1]
c_max = np.maximum(Covertype_lewis_prob / Covertype_l2_lp_prob, Covertype_l2_lp_prob / Covertype_lewis_prob)
Covertype_c = max(c_max)
print("Covertype_c:", Covertype_c)
print("")


index_max = np.argmax(c_max)
print("Das Maximum von c befindet sich am Index:", index_max)
print("Der Wert von c am Index:", c_max[index_max])
print("Covertype_lewis_prob[index_max]:", Covertype_lewis_prob[index_max])
print("Covertype_l2_lp_prob[index_max]:", Covertype_l2_lp_prob[index_max])
print("")

print("Nach der Untersuchung ist bei Covertype die Scores von lp und l2 unterschiedlich, obwohl p=2 gilt.")
print("")

test_lp = compute_leverage_scores(Covertype.X, p=2)
test_l2 = calculate_lev_score_exact(Covertype.X)
print("Rang von X:", Covertype.X.shape[1])
print("Summe von lp-Scores:", sum(test_lp))
print("Summe von l2-Scores:", sum(test_l2))
print("Hier sieht man, dass die Summe von l2-Scores nicht gleich dem vollen Rang von X ist. ")
print("")
```
