```python
import numpy as np
%store -r Iris

from efficient_probit_regression.sampling import compute_leverage_scores, calculate_lewis_weights_exact, to_density, calculate_l2_lp_leverage_score

p = 2.0

leverage_score = compute_leverage_scores(Iris.get_X(),p,False)
# print(leverage_score)
print("Rank of X", np.linalg.matrix_rank(Iris.get_X()))
print("Leverage_score sum:", sum(leverage_score), "p =", p)
# print("density:", to_density(leverage_score))
print("Density of leverage_score sum:", to_density(leverage_score).sum())

lewis_weight = calculate_lewis_weights_exact(Iris.get_X(), p, 20)
# print(lewis_weight)
print("Lewis_weight sum:", sum(lewis_weight), "p =", p)

l2_lp_leverage_score = calculate_l2_lp_leverage_score(Iris.get_X(), p, False)[0]
l2_lp_p = calculate_l2_lp_leverage_score(Iris.get_X(), p, False)[1]
# print("The combination of l_2 and l_p scores:", l2_lp_leverage_score)
print("The sum of l_2 and l_p scores:", l2_lp_leverage_score.sum())
# print("The density of l_2 and l_p scores:", l2_lp_p)
print("The sum of density from l_2 and l_p scores:", l2_lp_p.sum())
```
