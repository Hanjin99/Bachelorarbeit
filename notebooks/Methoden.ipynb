```python
import numpy as np
%store -r Iris

from efficient_probit_regression.sampling import compute_leverage_scores, calculate_lewis_weights_exact, to_density

from efficient_probit_regression.total_variation_distance import total_variation_distance

p = 2.0

leverage_score = compute_leverage_scores(Iris.get_X(),p,False)
lewis_weight = calculate_lewis_weights_exact(Iris.get_X(), p, 20)

prob_leverage_score = to_density(leverage_score)
# print(prob_leverage_score)
# print("Prob Leverage Score sum:", prob_leverage_score.sum())

prob_lewis_weight = to_density(lewis_weight)
# print(prob_lewis_weight)
# print("Prob Lewis Weights sum:", prob_lewis_weight.sum())

print("TVD from Leverage Score and Lewis Weights:")

tvd_lev_lewis = total_variation_distance(prob_leverage_score, prob_lewis_weight)
print("TVD via total_variation_distance() with normalize = False:")
print(tvd_lev_lewis,", for p =", p)

tvd_lev_lewis_normalize = total_variation_distance(leverage_score, lewis_weight, normalize=True)
print("TVD via total_variation_distance() with normalize = True:")
print(tvd_lev_lewis_normalize,", for p =", p)

tvd_manuel = 0.5 * np.abs(prob_leverage_score - prob_lewis_weight).sum()
print("TVD via 0.5 * np.abs(prob_leverage_score - prob_lewis_weight).sum():")
print(tvd_manuel,", for p =", p)

p = 1.0

leverage_score = compute_leverage_scores(Iris.get_X(),p,False)
lewis_weight = calculate_lewis_weights_exact(Iris.get_X(), p, 20)
prob_leverage_score = to_density(leverage_score)
prob_lewis_weight = to_density(lewis_weight)

tvd_lev_lewis = total_variation_distance(prob_leverage_score, prob_lewis_weight)
print("TVD via total_variation_distance() with normalize = False:")
print(tvd_lev_lewis,", for p =", p)

tvd_lev_lewis_normalize = total_variation_distance(leverage_score, lewis_weight, normalize=True)
print("TVD via total_variation_distance() with normalize = True:")
print(tvd_lev_lewis_normalize,", for p =", p)

tvd_manuel = 0.5 * np.abs(prob_leverage_score - prob_lewis_weight).sum()
print("TVD via 0.5 * np.abs(prob_leverage_score - prob_lewis_weight).sum():")
print(tvd_manuel,", for p =", p)

```
